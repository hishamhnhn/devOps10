```python
import pandas as pd
df = pd.read_csv('ML-chat-reg.csv')
df
```

```python
df['gender'] = df['gender'].replace('Male',1)
df['gender'] = df['gender'].replace('Female',0)
```

```python
df
```

```python
x = df [['age','gender']]
y = df['likeness']
```

```python
from sklearn.neighbors import KNeighborsClassifier

#training the modelusing given data set 

model = KNeighborsClassifier(n_neighbors=3)

#fitting the model  

model.fit(x,y)

predicted = model.predict([[33,56]]) # again it 2d array needs two [[x]]

predicted
```

```python
model.predict(x)
```

metrice for evaluation 

## split data into test and train (80/20)

```python
from sklearn.model_selection import train_test_split # importing the lib for data spliting 
from sklearn.metrics import accuracy_score   #importing the lib for metrice measurment 

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

#create model a model 
model = KNeighborsClassifier()
#fitting the model
model.fit(x_train,y_train)

#prediction 

predicted = model.predict(x_test)


score = accuracy_score(y_test,predicted) 
score

print('score of the model is',score)
```
