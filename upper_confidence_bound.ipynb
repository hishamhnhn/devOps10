# Upper Confidence Bound (UCB)

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing UCB


```python
numbers_of_selections = [0]*d  # d is number of rows 
sums_of_rewards = [0]*d
total_reward = 0 
import math
N = 10000
d = 10
ads_selected = []
for n in range(0,N): # total number of index form 0 to 10000 
    max_upper_bound = 0
    ad = 0 
    for i in range(0,d): # this is the each vesion of the add 
        if (numbers_of_selections[i]>0):
            average_reward = sums_of_rewards[i]/numbers_of_selections[i]
            delta_i = math.sqrt(3/2*math.log(n+1)/numbers_of_selections[i])
            upper_bound = average_reward+delta_i
        else:
            upper_bound = 1e400 # upper bound to high 
        if  upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = i
            
    ads_selected.append(ad)
    numbers_of_selections[ad] = numbers_of_selections[ad] = i
    reward = dataset.values[n,ad]
    sums_of_rewards[ad] = sums_of_rewards[ad] + reward
    total_reward = total_reward + reward
        
        
            
    
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
