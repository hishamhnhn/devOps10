```python
#import dataset
from sklearn import datasets
#load dataset 

canser = datasets.load_breast_cancer()
```

```python
#print features in data set 
# the data has 30 features 
print('features',canser.feature_names)

# the dataset has 2 labels 
print('labels',canser.target_names)
```

```python
# the dataset has 569 rows and 30 colums as shown below 
canser.data.shape
```

```python
# printing the first row of the dataset 
print(canser.data[0:1])
```

```python
print(canser.target)
```

### spliting the dataset 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(canser.data,canser.target,test_size=0.2,random_state=0)
```

### creating the SVM model 

```python
from sklearn import svm

# create a SVM classifier for with linear kernel
clf = svm.SVC(kernel='linear')

clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)

y_pred
```

### check the acurracy 

```python
# check accuracy 
from sklearn import metrics

score = metrics.accuracy_score(y_test,y_pred)

print('test results:',score*100)
```

### check the confusion matrix 

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
cm
```
