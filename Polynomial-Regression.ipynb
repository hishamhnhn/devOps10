```python
#### bad fitting example or non fitted example 


import numpy as np
import matplotlib.pyplot as plt 
import random

x = random.sample(range(100), 50) 
y = random.sample(range(100), 50)

mymodel = np.poly1d(np.polyfit(x,y,3))

myline = numpy.linspace(2,95,100)

plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

## R square for bad fit 

```python
import numpy  # numpy can also used as resouce for ML algo
from sklearn.metrics import r2_score

model = numpy.poly1d(numpy.polyfit(x,y,3)) #ploy 1d line 
                                
myline = numpy.linspace(0,100,50)

plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()


print(r2_score(y,model(x)))
```
