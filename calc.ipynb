```python

def calc_1():
    a = int(input('enter a = '))
    opt = input('enter operator ')
    b = int(input('enter b = '))
    if opt == '+':
        c = a+b
    elif opt == '-':
        c = a-b
    elif opt == '/':
        c = a/b
    elif opt == '*':
        c = a*b
    elif opt == '%':
        c = a%b
    return print('a '+ opt+' b = ',c)
```
